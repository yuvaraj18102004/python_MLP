```python
# Write a Python program that takes the length and width of a rectangle from the user and prints its area.

w=float(input("Enter the width of the rectangle : "))
h=float(input("Enter the height of the rectangle : "))
r=float(input("Enter the radius of the circle : "))

print("The area of rectangle is : ",w*h)
print("The area of circle is : ",3.14*r*r)
```
