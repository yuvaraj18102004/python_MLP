```python
#Write a program that takes the purchase amount as input and calculates the discount and final amount to be paid.

s=float(input("Enter the amount : "))

if(s>=100 and s<=500):
    print("your total amount is :",(s*10/100)-s)
elif(s>500):
    print("your total amount is :",(s*20/100)-s)
elif(s<100):
    print("you not have any discount")
else:
    print("Tnank you")
    
```

```python
#Create a program that simulates a traffic light. The user inputs one of the colors: Red, Yellow, or Green. Based on the input

colour=str(input("Enter the colour :"))

if(colour=="red"):
    print("Stop")
elif(colour=="yellow"):
    print("Ready to go ")
elif(colour=="green"):
    print("Go")
else:
    print("invalid colour")
```

```python
#Write a program that accepts the student's marks and displays their grade

marks=int(input("Enter your mark :"))
if(marks>=90):
    print("Grade A")
elif(marks>=75 and marks<90):
    print("Grade B")
elif(marks>=50 and marks<75):
    print("Grade C")
elif(marks<50):
    print("fail")
else:
    print("invalid marks")
```

```python
#Write a program that checks the strength of a password based on these rules

password=str(input("Enter the password :"))
if(len(password)>=8):
    print("Strong")
elif(len(password)>5 or len(password)<7):
    print("Medium")
elif(len(password)<5):
    print("Week")
else:
    print("invalid password")
```
